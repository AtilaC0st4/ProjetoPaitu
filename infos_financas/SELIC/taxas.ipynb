```python
import pandas as pd
import matplotlib.pyplot as plt

def CarregaSELIC(data_inicial,data_final):

   # carrega um intevalor de dados do bcb

   url_bcb = f"https://api.bcb.gov.br/dados/serie/bcdata.sgs.4189/dados?formato=csv&dataInicial={data_inicial}&dataFinal={data_final}"
               
   serie_SELIC = pd.read_csv(url_bcb, sep=";")
   return serie_SELIC

   # Abaixo come√ßa o codigo
   serie = CarregaSELIC("01/01/2023","01/02/2023")

   serie1 = serie.replace({',':'.'}, regex=True) 

   serie1['valor'] = serie1['valor'].astype(float)

   serie1.plot.line (x="data",y="valor",title="SELIC taxas ao ano",legend = True) 
   plot.show(block=True)
   plt.show()

```
