```python
import requests

link = "https://olinda.bcb.gov.br/olinda/servico/selic_contas/versao/v1/odata/QuantidadePorTipoConta(Data=@Data)?@Data=%272019-02-02%27&$top=1000&$format=json&$select=Data,Tipo,NomeTipoFuncaoConta,NomeDestinacaoConta,Quantidade"
Requesições = requests.get(link)
informacoes = Requesições.json()

print(informacoes)
```

```python
import pandas as pd

Tabela = pd.DataFrame(informacoes["value"])
display(Tabela)
```
